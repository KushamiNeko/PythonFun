```python
import pandas as pd
import os
```

```python
# folder = "IPO With Sales Growth"
# folder = "Top Industries With Sales Growth"
folder = "Earnings + Sales Growth"

root = os.path.join("data", folder)

df = None

for f in os.listdir(root):
    file = os.path.join(root, f)

    dataframe = pd.read_csv(file)

    df = dataframe if df is None else df.append(dataframe)

df = df.reset_index(drop=True)

df
```

```python
filter = df["Sector"].isna()

df = df[~filter]

filter = df.loc[:, "Sector"].transform(lambda x: x.strip()) == ""

df = df[~filter]

df
```

```python
df = df.reset_index(drop=True)
df
```

```python
output = ",".join(df["Ticker"])

output
```

```python
with open(f"export/{folder}/output.txt", "w") as f:
    f.write(output)
```
